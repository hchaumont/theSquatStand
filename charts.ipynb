```python
import numpy as np
import pandas as pd
```

```python
from datetime import datetime, date
```

```python
date.today()
```

```python
thisWeek = pd.date_range(date.today(), periods=7)
```

```python
thisWeek
```

```python
volume = [5000, 7000, 9000, 11000, 6000, 8000, 10000]
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
output_notebook()
```

```python
p = figure(title="Daily Volume", x_axis_label='date', y_axis_label='volume', x_axis_type='datetime')
```

```python
p.line(x=thisWeek, y=volume, line_width=3)
```

```python
show(p)
```

```python
week = [str(d).split()[0] for d in thisWeek]
```

```python
q = figure(title="Daily Volume", x_range=week, y_axis_label='volume')
```

```python
q.vbar(x=week, top=volume, width=.9)
```

```python
show(q)
```
